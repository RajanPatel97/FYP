```python
import csv
import cv2

str_return = 'First image returned None, should be valid, maybe the script is not in the correct path'
files_new = {}

### Quick check to test that the filepath from the csv
### points to a valid image
test_im = 0

with open('new_info.csv', 'r', encoding="utf8") as file:
    f = csv.reader(file, delimiter = '|')
    header = next(f)
    for i, line in enumerate(f):
        files_new[line[0]] = True
        ### Try to open the first image
        if test_im == 0:
            img = cv2.imread(line[0])
            assert img is not None, str_return
            test_im = 1
import os

with open('info.csv', 'r', encoding="utf8") as file:
    f = csv.reader(file, delimiter = '|')
    header = next(f)
    for i, line in enumerate(f):
        if line[0] not in files_new:
            if os.path.exists(line[0]):
                os.remove(line[0])
                print('File {:s} removed'.format(line[0]))
```
